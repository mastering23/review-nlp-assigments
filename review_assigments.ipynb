```python
# RUN THIS CELL to perform standard imports :
import spacy
nlp = spacy.load('en_core_web_sm')
```

 1. Create a Doc object from the file owlcreek.txt

```python
# Enter your code here: 
with open('../Documents/owlcreek.txt') as f:
    text = f.read()
    print(text)
    
```

2. How many tokens are contained in the file? 

```python
doc = nlp(text)
```

```python
num_tokens = len(doc)
print(f"Number of tokens: {num_tokens}")
```

3. How many sentecens are contained in the file?

```python
num_sentences = len(list(doc.sents))
print(f"Number of sentences: {num_sentences}")
```

4. Print the second sentece in the document

```python
sentences = list(doc.sents)
print(sentences[1].text)
```

5. For each token in the sentence above, print its text, POS tag, DEP tag and lemma
CHALLENGE: Have values line up in columns in the print output.

```python
second_sentence = sentences[1]
print(f"{'TEXT':<15}{'POS':<10}{'DEP':<15}{'LEMMA':<15}")
print("-" * 55)
for token in second_sentence:
    print(f"{token.text:<15}{token.pos_:<10}{token.dep_:<15}{token.lemma_:<15}")
```

6. Write a matcher called 'Swimming' that finds both occurrences of the phrase "swimming vigorously" in the text"

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)
```

```python
pattern = [
    {'LOWER': 'swimming'},
    {'IS_SPACE': True},
    {'LOWER': 'vigorously'}
]
```

```python
matcher.add('Swimming', [pattern])
```

```python
doc = nlp(text)
```

```python
matches = matcher(doc)
```

```python
for match_id, start, end in matches:
    span = doc[start:end]
    print(span.text)
```
